```python
import pandas as pd
import numpy as np
```

```python
business_license = pd.read_csv('2016_Active_Business_License_Data.csv')
```

```python
#find business license start from 2016
index_list = []
for i in range(len(business_license)):
    if business_license['License Start Date'][i].split('-')[-1] == '16':
        index_list.append(i)
        
business_license_2016 = business_license.iloc[index_list].reset_index(drop = True)
```

```python
zipcode_list = []
for i in range(len(business_license_2016)):
    zipcode = business_license_2016['City, State, ZIP'][i].split(' ')[-1].split('-')[0]
    zipcode_list.append(zipcode)

business_license_2016['Zipcode'] = zipcode_list
```

```python
business_license_2016.to_csv('aaa.csv')
```
