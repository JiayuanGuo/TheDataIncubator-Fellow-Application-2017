```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

## Housing Building Permits 2014 to 2016

```python
bp = pd.read_csv('bp.csv')
bp_2014 = bp[bp['Expected Construct Year'] == 2014].reset_index(drop = True)
bp_2015 = bp[bp['Expected Construct Year'] == 2015].reset_index(drop = True)
bp_2016 = bp[bp['Expected Construct Year'] == 2016].reset_index(drop = True)
```

```python
#2016
#get zipcode list and corresponding counts
zipcode_list_16 = []
for i in range(len(bp_2016)):
    zipcode = bp_2016['Zipcode'][i]
    if zipcode not in zipcode_list_16:
        zipcode_list_16.append(zipcode)
        
df_16 = pd.DataFrame()
df_16['Zipcode'] = zipcode_list_16
category = ['MULTIFAMILY','SINGLE FAMILY / DUPLEX', 'COMMERCIAL']

for i in category:
    bp_cat = bp_2016[bp_2016['Category'] == i]
    cat_counts_list = []
    for j in zipcode_list_16:
        cat_counts = len(bp_cat[bp_cat['Zipcode'] == j])
        cat_counts_list.append(cat_counts)
    df_16['%s' % i] = cat_counts_list
    
df_16['HOUSING'] = df_16['MULTIFAMILY'] + df_16['SINGLE FAMILY / DUPLEX']
df_16.sort_values(['HOUSING']).tail(10)
```
