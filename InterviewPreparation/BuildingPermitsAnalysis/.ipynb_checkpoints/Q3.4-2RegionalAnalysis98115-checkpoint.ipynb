```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

# Commercial Building License

```python
#building license
business_license = pd.read_csv('2016_Active_Business_License_Data.csv')
business_license.drop('NAICS Code', axis=1, inplace=True)
business_license.drop('Ownership Type', axis=1, inplace=True)
```

```python
#add zipcode
zipcode_list= []
for i in range(len(business_license)):
    zipcode = business_license['City, State, ZIP'][i].split(' ')[-1].split('-')[0]
    zipcode_list.append(zipcode)

business_license['Zipcode'] = zipcode_list
```

```python
#add year and month
year_list = []
month_list = []
for i in range(len(business_license)):
    year= business_license['License Start Date'][i].split('-')[-1]
    month = business_license['License Start Date'][i].split('-')[1]
    year_list.append(year)
    month_list.append(month)
business_license['Year'] = year_list
business_license['Month'] = month_list
#delete useless columns
business_license.drop(['License Start Date','City, State, ZIP'], axis=1, inplace=True)
```

```python
business_license
```

## For 98115

```python
bl_98115 = business_license[business_license['Zipcode'] == '98115'].reset_index(drop = True)
```

```python
#find business license start from 2016
bl_98115_16 = bl_98115[bl_98115['Year'] == '16'].reset_index(drop = True)

#create counting df
count_df_16 = pd.DataFrame()
count_df_16['Year'] = [2016] *12
count_df_16['Month'] = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
month_count_16 = []
for i in count_df_16['Month']:
    count = len(bl_98115_16[bl_98115_16['Month'] == i])
    month_count_16.append(count)
count_df_16['Count'] = month_count_16
```

```python
#find business license start from 2015
bl_98115_15 = bl_98115[bl_98115['Year'] == '15'].reset_index(drop = True)

#create counting df
count_df_15 = pd.DataFrame()
count_df_15['Year'] = [2015] *12
count_df_15['Month'] = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
month_count_15 = []
for i in count_df_15['Month']:
    count = len(bl_98115_15[bl_98115_15['Month'] == i])
    month_count_15.append(count)
count_df_15['Count'] = month_count_15
```

```python
#find business license start from 2014
bl_98115_14 = bl_98115[bl_98115['Year'] == '14'].reset_index(drop = True)

#create counting df
count_df_14 = pd.DataFrame()
count_df_14['Year'] = [2014] *12
count_df_14['Month'] = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
month_count_14 = []
for i in count_df_14['Month']:
    count = len(bl_98115_14[bl_98115_14['Month'] == i])
    month_count_14.append(count)
count_df_14['Count'] = month_count_14
```

```python
count_df = pd.concat([count_df_14,count_df_15,count_df_16]).reset_index(drop = True)
```

```python
count_df
```

# Housing Building Permits

```python
bp = pd.read_csv('bp.csv')
bp_98115 = bp[bp['Zipcode'] == '98115'].reset_index(drop = True)
bp_98115['Category'].replace(['SINGLE FAMILY / DUPLEX', 'MULTIFAMILY'], 'HOUSING',inplace = True)
bp_98115_housing = bp_98115[bp_98115['Category'] == 'HOUSING'].reset_index(drop = True)
```

```python
#2014-2016 housing count
count_housing = []
for j in [2014,2015,2016]:
    for i in range(12):
        count = len(bp_98115_housing[(bp_98115_housing['Expected Construct Year'] == j) & 
                     (bp_98115_housing['Expected Construct Month'] == i+1)])
        count_housing.append(count)
count_housing
```

```python
#2014-2016 commercial count
bp_98115_com = bp_98115[bp_98115['Category'] == 'COMMERCIAL'].reset_index(drop = True)

count_commercial = []
for j in [2014,2015,2016]:
    for i in range(12):
        count = len(bp_98115_com[(bp_98115_com['Expected Construct Year'] == j) & 
                     (bp_98115_com['Expected Construct Month'] == i+1)])
        count_commercial.append(count)
count_commercial
```

```python
count_df['HOUSING'] = count_housing
```

```python
count_df['COMMERCIAL'] = count_commercial
```

```python
count_df.to_csv('count.csv')
```
